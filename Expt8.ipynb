```python
# NumPy--> Numerical Python
## Creates arrays and performs numerical operations, faster than lists
```

```python
# Importing numpy and forming arrays
import numpy as snake ## generally imported as 'np' however you can use anything

p = snake.array([10, 20, 30, 40])
q = snake.array([[1, 2, 3],[4, 5, 6]])
print(p)
print(q)
```

```python
#Dimension of p
p.ndim
```

```python
# Shape of a-->syntax:(rows, columns)
p.shape
```

```python
# Shape of b
q.shape
```

```python
#datatype-->here, 64 is number of bits
p.dtype
```

```python
# Formation of a zero matrix
snake.zeros((2,3))
```

```python
# Formation of a matrix containing only ones
snake.ones((3,3))
```

```python
# Diagonal matrix--> needs to be symmetric hence input only one number
snake.eye((3)) ##matrix size is 3x3
```

```python
# "ARANGE" function
snake.arange(1,10,2)
```

```python
#"linspace" command
snake.linspace(0,1,5)
```

```python
#Basic operations-->original array does not get modified
a= p + 5
b= p*3
print(a)
print(b)
```

```python
#Mean command-->average
snake.mean(p) ##value is 25.0
```

```python
#Median command
snake.median(p)  ##here, median also 25.0

##if mean = median then your data does not have any outliers and is perfect however a vast difference in mean and median means an imperfect data
## median is middle value and since here number of elements is even hence value comes out to 25, if it was odd then we would get middle number
```

```python
# Max command
snake.max(q)
```

```python
#Min command
snake.min(q)
```

```python
#Sum of array
snake.sum(p)
```
